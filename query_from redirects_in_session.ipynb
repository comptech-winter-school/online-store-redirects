```python
with open("420_searches.json") as ff:
    data = json.load(ff)
```

```python
query = []
product_external_id = []
for i in range(len(data)):
    query.append(list(data[i]['products'].keys()))
    product_external_id.append(data[i]['query'])
    i+=1
```

```python
searches = pd.DataFrame({'query': product_external_id, 'external_id': query})
searches.head(3)
```

```python
query_from_redirects_in_searches = searches[searches['query'].isin(redirects['query'])].reset_index()
query_from_redirects_in_searches.head(3)
```

```python
query_from_redirects_in_searches = query_from_redirects_in_searches[['query', 'external_id']].explode('external_id').reset_index()
query_from_redirects_in_searches = query_from_redirects_in_searches[['query', 'external_id']]
query_from_redirects_in_searches.head(3)
```

```python
query_from_redirects_in_searches['query'].nunique()
```

В таблице с сессиями '420_searches.json'  так же два одинаковых запроса, которые встречаются в таблице с положительными редиректами.

```python
query_from_redirects_in_searches.to_csv('query_from_redirects_in_session.csv')
```
