#### Таблица из общих запросов (запрос содержится среди редиректов и сессий)

```python
import pandas as pd
import json
from pprint import pprint
```

```python
redirects = pd.read_csv("420_redirects.csv", index_col=0)
categories = pd.read_csv("categories.csv", index_col=0)
products = pd.read_csv("products.csv", index_col=0)
products_categories = pd.read_csv("products_categories.csv", index_col=0)
```

```python
with open("420_searches.json") as ff:
    data = json.load(ff)
```

```python
pprint(data[0])
```

```python
query = []
product_external_id = []
amount_click = []
for i in range(len(data)):
    query.append(list(data[i]['products'].keys()))
    product_external_id.append(data[i]['query'])
    amount_click.append(list(data[0]['products'].values())) 
    i+=1
```

```python
searches = pd.DataFrame({'query': product_external_id, 'external_id': query, 'amount_click': amount_click})
searches.head(3)
```

```python
from_redirects_in_searches = searches[searches['query'].isin(redirects['query'])].reset_index()
from_redirects_in_searches
```

```python
from_redirects_in_searches_1 = from_redirects_in_searches[['query', 'external_id', 'amount_click']]\
       .explode('external_id', 'amount_click').reset_index()
from_redirects_in_searches_1 = from_redirects_in_searches_1[['query', 'external_id', 'amount_click']]\
       .explode('amount_click').reset_index()
from_redirects_in_searches_1 = from_redirects_in_searches_1[['query', 'external_id', 'amount_click']]
from_redirects_in_searches_1.head(3)
```

```python
from_redirects_in_searches_1 = from_redirects_in_searches_1[from_redirects_in_searches_1['external_id'] != '14078700020-1']
```

```python
from_redirects_in_searches_1['external_id'] = pd.to_numeric(from_redirects_in_searches_1['external_id'])
from_redirects_in_searches_1['amount_click'] = pd.to_numeric(from_redirects_in_searches_1['amount_click'])
```

```python
from_redirects_in_searches_1 = pd.merge(from_redirects_in_searches_1, products[['name', 'external_id', 'id']], on='external_id')
from_redirects_in_searches_1.head(3)
```

```python
from_redirects_in_searches_1 = pd.merge(from_redirects_in_searches_1, products_categories[['category_id', 'product_id']], left_on='id', right_on='product_id')
from_redirects_in_searches_1.head(3)
```

```python
from_redirects_in_searches_1 = pd.merge(from_redirects_in_searches_1, redirects[['query', 'category_id', 'category']],\
                                        on='query')
```

```python
del from_redirects_in_searches_1['external_id']
del from_redirects_in_searches_1['id']
```

```python
from_redirects_in_searches_1.drop_duplicates(keep='first', inplace=True)
from_redirects_in_searches_1.columns = ['query', 'amount_click', 'name', 'category_id', 
                                       'product_id','category_id_redirect', 'category']
```
