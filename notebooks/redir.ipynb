```python
categories.dropna(inplace=True)
```

```python
categories.parent_id = categories.parent_id.astype('int64')
```

```python
categories.parent_id.nunique()
```

```python
categories.head()
```

```python
categories_by_parent = categories.groupby('parent_id')['id'].apply(list)
```

```python
categories_by_parent
```

```python
products_by_categories
```

```python
selected_categories = products_by_categories.index.tolist()
```

```python
len(selected_categories)
```

```python
parent_list = categories_by_parent.index.tolist()
```

```python
len(parent_list)
```

```python
#словарь: ключи - родительские категории
category_tree = dict.fromkeys(parent_list)
```

```python
for name in category_tree.keys():
    category_tree[name] = dict.fromkeys(categories_by_parent[name])
    for subname in category_tree[name].keys():
        if subname in selected_categories:
            category_tree[name][subname] = products_by_categories[subname]
```

Структура дерева: list_of_products = category_tree[parent][category]
